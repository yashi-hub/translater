```python
#Import All the libraries including pandas for using multiple functions.
from tqdm import tqdm
import pandas as pd
from collections import Counter
```

```python
#Reading the words from the Dictionary given.
df=pd.read_csv('french_dictionary.csv')
```

```python
#List Of words of English and French.
df
```

```python
dic={}
for i in range(1000):
    dic[df.iloc[i,0]]=df.iloc[i,1]
```

```python
#Number Of words in the dictionary.
len(dic)
```

```python
input_file=open('t8.shakespeare.txt')
```

```python
list_of_english_words_replaced=[]
```

```python
#Time taken to process in Iterations/second.
g=open('check.txt','w')
for line in tqdm(input_file):
    for word in line.split():
        try:
            g.write(dic[word.lower()])
            g.write(" ")
            list_of_english_words_replaced.append(word)
        except:
            g.write(word)
            g.write(" ")
    g.write("\n")
g.close()
```

```python
df_words=pd.DataFrame()
df_words['english word replaced']=list_of_english_words_replaced
```

```python
#Unique Number of words that was replaced with French words from the dictionary.
df_words['english word replaced']=df_words['english word replaced'].apply(lambda x:x.lower())
len(df_words['english word replaced'].unique())
```

```python
#Unique list of words that was replaced with French words from the dictionary.
#Number of times a word was replace or Frequency of each word replaced.
Counter(df_words['english word replaced'])
```
